```python
def read_data(filename):
    temp = pd.read_excel(filename)

    colname = re.search(r'(.*?)市场.*',filename).group(1)
    if temp.时间.dtype == 'int64':
        temp['时间'] = pd.to_datetime(temp['时间'],unit = 'D', origin = '1899-12-30')

    temp.rename(columns = {'交易金额':colname}, inplace = True)#把原来名字叫交易金额的改成colname

    temp.set_index(keys = '时间', inplace = True)#keys指定了要变成索引的列
    return temp
```

```python
def process_data(df, by):
    predict_amount = df.groupby(by=by)[['预估销售额']].sum()
    predict_amount['预估销售额占比'] = predict_amount['预估销售额'] / predict_amount.sum().values
    # 计算宝贝数
    predict_amount['宝贝数'] = df.drop_duplicates(subset=['宝贝ID']).groupby(by=by)['宝贝ID'].count()
    predict_amount['宝贝数占比'] = predict_amount['宝贝数'] / predict_amount['宝贝数'].sum()
    predict_amount['单宝贝平均销售额'] = predict_amount['预估销售额'] / predict_amount['宝贝数']
    predict_amount['相对竞争度'] = 1 - ((predict_amount['单宝贝平均销售额'] - predict_amount['单宝贝平均销售额'].min()) / (predict_amount['单宝贝平均销售额'].max() - predict_amount['单宝贝平均销售额'].min()))
    # 按照相对竞争度的顺序来排序
    predict_amount.sort_values(by='相对竞争度', inplace=True)
    return predict_amount
```

```python
def gen_sale_data(df):
    sale = df.groupby(by='商品')[['交易金额']].sum()
    sale['交易金额占比']=sale['交易金额'] / sale['交易金额'].sum()
    sale['商品个数']=df.groupby(by='商品')['商品'].count()
    sale['交易增长幅度']=df.groupby(by='商品')['交易增长幅度'].mean()
    #reset_index是重置索引的操作
    sale.reset_index(inplace=True)
    return sale
```

```python
#对交易金额和交易增长幅度进行盖帽处理
#盖帽一般使用90%或者80%的值去盖帽
#计算交易金额列的第90百分位数
#threshold为阈值，界值，起始点
threshold = np.percentile(bayer_sale['交易金额'],90)
```

```python
def mask(s):
    threshold = s.quantile(q=0.9)
    cond = s >= threshold
    s = s.mask(cond,threshold)
    return s
```

```python
def get_star_cow_ques_product(df):
    star = df.loc[(df['交易金额占比'] >=df['交易金额占比'].quantile(0.5)) & (df['交易增长幅度'] >=df['交易增长幅度'].quantile(0.5))]
    cow = df.loc[(df['交易金额占比'] >=df['交易金额占比'].quantile(0.5)) & (df['交易增长幅度'] <df['交易增长幅度'].quantile(0.5))]
    ques = df.loc[(df['交易金额占比'] < df['交易金额占比'].quantile(0.5)) & (df['交易增长幅度'] >=df['交易增长幅度'].quantile(0.5))]
    return star, cow,ques
```

```python
def flow_analyse(df,name):
    #根据交易指数降序排序
    df.sort_values(by = '交易指数',ascending=False,inplace=True)
    #以 “流量来源” 这一列作为索引，接着提取出 “交易指数” 这一列的数据
    df_flow = df.head(10).set_index(keys = '流量来源')['交易指数']
    cond = np.array([True if index in paid else False for index in df_flow.index])
    df_flow.plot(kind='pie',autopct='%.2f%%',figsize=(7,7),explode=cond*0.1)
    plt.title(name)
    total_flow=df_flow.sum()
    total_paid_flow = df_flow[cond].sum()
    #radio是比率的意思
    total_paid_flow_ratio = total_paid_flow / total_flow
    #以百分数形式呈现，且保留两位小数
    print(f'在前10的流量中，总交易指数:{total_flow},付费流量占比{total_paid_flow_ratio * 100:.2}%,付费流量带来的交易指数：{total_paid_flow}')
```
