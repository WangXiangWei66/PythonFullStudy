### 2. 广播机制

【重要】ndarray广播机制的两条规则
- 规则一：为缺失的维度补1
- 规则二：假定缺失元素用已有值填充


例1：
m = np.ones((2, 3))
a = np.arange(3)
求M+a

```python
# 就是为了让两个进行计算的ndarray形状相同, 进而可以进行计算. 
m = np.ones((2, 3))
m
```

```python
a = np.arange(3)
a
```

```python
m + a
```

```python
a [0, 1, 2]  -> 加个维度 --> [[0, 1, 2]], shape:(1, 3) ---> 缺数据, 用已有的数据填充 --> [[0, 1, 2], [0, 1, 2]], shape:(2, 3)
```

例2：
a = np.arange(3).reshape((3, 1))
b = np.arange(3)
求a+b

```python
a = np.arange(3).reshape((3, 1))
a
```

```python
b = np.arange(3)
b
```

```python
a + b
```

```python
# 冒泡排序: 
# 原理: 相邻的元素进行比较, 每次比较一轮, 最大的那个元素, 会跑后面. 
# 时间复杂度: 2层循环, 所以是O(n^2)
# 空间复杂度: 没有产生额外的空间来存储, O(1)
# 稳定性: 稳定的. 

def bubble_sort(arr):
    length = len(arr)
    # 最后一个元素, 不用循环, 因为后面没有可以比较的元素了. 
    for i in range(length - 1):
        # 小优化
        switched = False
        # 外层循环的次数减去排序次数
        for j in range(length - i - 1):
            # 相邻元素进行比较
            if arr[j] > arr[j + 1]:
                # 顺序不对, 要交换位置. 
                # 不使用第三个变量, 最少用三种方法, 交换2个变量的值. 
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                # 如果没发生交换, 说明数组本身是有序的. 后面就不用比了. 
                switched = True
        if not switched:
            break
    return arr
```

```python
l = [6, 1, 2, 5, 4, 3]
bubble_sort(l)
```

```python
# 选择排序   
# 原理: 从未排序的数组中, 选择一个最小的, 放到最前面. 
# 时间复杂度: O(n^2)
# 空间复杂度: O(1)
# 稳定性: 不稳定

def select_sort(arr):
    length = len(arr)
    for i in range(length -1):
        # 设置最小值
        least = i
        for j in range(i + 1, length):
            if arr[least] > arr[j]:
                # 更新最小值的索引
                least = j
        # 交换位置
        arr[least], arr[i] = arr[i], arr[least]
        
    return arr
```

```python
l = [6, 1, 2, 5, 4, 3]
select_sort(l)
```

```python
# 快速排序
#  原理: 选择一个基准, 所有比基准小的, 放到基准前面, 比基准大的放到基准后面. 重复这个过程, 直到数组中只有一个元素
# 时间复杂度: O(nlog2(n))
# 空间复杂度: O(nlog2(n))
# 稳定性: 不稳定

def quick_sort(arr):
    # 只有一个元素, 就不用排了, 直接返回. 
    # 给递归一个返回的条件. 
    if len(arr) < 2:
        return arr
    
    # 选择基准值
    pivot =  arr.pop()
    greater = []
    less = []
    
    for i in range(len(arr)):
        if arr[i] > pivot:
            greater.append(arr[i])
        else:
            less.append(arr[i])
            
    # 递归调用, 递归就是自己调自己
    return quick_sort(less) + [pivot] + quick_sort(greater)
```

```python
l = [6, 1, 2, 5, 4, 3]
quick_sort(l)
```

```python
# 插入排序
# 原理: 认为第一个元素是有序的, 遍历后面的元素, 和前面的元素对比, 找到合适的位置插入. 
# 时间复杂度: O(n^2)
# 空间复杂度: O(1)
# 稳定性: 稳定

def insert_sort(arr):
    for i in range(1, len(arr)):
        loop_index = i
        while loop_index > 0 and arr[loop_index] < arr[loop_index - 1]:
            arr[loop_index], arr[loop_index - 1] = arr[loop_index - 1], arr[loop_index]
            loop_index -= 1
    return arr
            
```

```python
l = [6, 1, 2, 5, 4, 3]
insert_sort(l)
```
