# 研究步长对梯度下降的影响

```python
x_origin = -9
learning_rate = 0.001
max_iter = 2000
threshold=1e-5
i = 0
ret = []
while True:#死循环，会不停的更新
    ret.append(x_origin)
    x_origin = x_origin - learning_rate * dx(x_origin)

    i+=1

    #下面设置退出条件
    #第一种
    # if i > max_iter:
    #     break

    #第二种
    # if abs(dx(x_origin)) < threshold:
    #     break

    #第三种
    if abs(x_origin - ret[-1]) < threshold:
        break
```
