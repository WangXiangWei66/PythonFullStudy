# 【关键词】最小二乘法，线性

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

$$\theta = (X^TX)^{-1}X^Ty$$

```python
x = np.linspace(0, 10, num = 20) + np.random.randn(20)
y = np.linspace(2, 8, num = 20) + np.random.randn(20)
plt.scatter(x, y)
```

$h_{\theta}$ = $\sum_{i=0}^{n}i$ $\theta_i$$x_i$   这个公式出来的前提是要在x之前加个x0，这一列x0全是1

```python
#-1在这里表示根据数组的元素总数自动推断这一维的长度，
X = np.hstack((np.ones(shape=(20,1)),x.reshape(-1,1)))
```

```python
X
```

```python
np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
```

```python
theta0 = 1.64266806 #截距
theta1 = 0.58884196 #系数
```

```python
plt.scatter(x,y)
x_test = np.linspace(-2,10,10)
y_test = theta1 * x_test + theta0
plt.plot(x_test, y_test, c='r')
```

### 抽取训练数据和预测数据

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 42)
```

```python
#研究每一个特征和标记之间的关系
#用每一个特征作为训练数据，建立他正与标记之间的关系
plt.figure(figsize = (2 *6, 4 * 5))
for i, col in enumerate(feature_names,1):
    #取数据
    X_train = df[[col]].values #获取col列的数据，并将它们转化为数组的形式返回
    linear = LinearRegression()
    linear.fit(X_train,target)
    #生成预测数据
    X_test = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1,1)
    y_=linear.predict(X_test)

    axes = plt.subplot(5,2,i)
    axes.plot(X_test,y_,c='r')
    axes.scatter(X_train, target)
    score = linear.score(X_train, target)
    axes.set_title(col + str(score))
```
